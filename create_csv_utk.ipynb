```python
from dotenv import load_dotenv
from pathlib import Path
import os
import pandas as pd

load_dotenv()
```

```python
file_dir = os.getenv("UTK_PATH")

def create_csv(file_dir):
    dir = os.listdir(file_dir)

    age = []
    gender = []
    race = []
    file_names = []
    for file in dir:
        file_name = file.split("_")

        file_names.append(file)

        try:
            age.append(int(file_name[0]))
            gender.append(int(file_name[1]))
            race.append(int(file_name[2]))
        except ValueError:
            print(f"Error: {file_name}")
            continue
    
    age = pd.Series(age, name="age")
    gender = pd.Series(gender, name="gender")
    race = pd.Series(race, name="race")
    file_names = pd.Series(file_names, name="filename")

    df = pd.concat([file_names, age, gender, race], axis=1).sample(frac=1, random_state=1).reset_index(drop=True)
    df.to_csv("utk.csv", index=False)
    

create_csv(file_dir)
```
