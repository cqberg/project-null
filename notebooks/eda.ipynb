```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("csv/utk.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df, x="age", kde=True)
plt.xlabel("Age")
plt.show()
```

```python
race_map = {
    0: "white",
    1: "black",
    2: "asian",
    3: "indian",
    4: "others"
}

df["race_label"] = df["race"].map(race_map)

plt.figure(figsize=(10,6))
sns.countplot(df, x="race_label")
plt.xlabel("Race")
plt.show()
```

```python
gender_count = df["gender"].value_counts()

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(df, x="gender", hue="gender", ax=ax[0])
ax[1].pie(gender_count, labels=["male", "female"], autopct="%1.1f%%")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(df, x="race_label", hue="gender")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df, x="age", gap=0.1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df, x="age", hue="gender", gap=0.1)
plt.tight_layout()
plt.show()
```
